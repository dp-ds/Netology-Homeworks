```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv( 'Mountains.csv', delimiter = ',' )
data.info()
```

```python
data.head(10)
```

```python
parent_mountains_count = data.groupby("Parent mountain").size()
parent_mountains_count.head(15)
```

```python
data_filter = parent_mountains_count[parent_mountains_count > 3]
data_filter
```

```python
data_filtered = data[data['Parent mountain'].isin(list(data_filter.index))]
data_filtered.shape
```

```python
data_filtered.index = range(len(data_filtered))
data_filtered.head(20)
```
